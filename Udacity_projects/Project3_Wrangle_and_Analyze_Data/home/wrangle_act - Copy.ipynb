```python
#Add the authenctication information to use the tweepy API to get the twitter data
consumer_key = 'XXXXXXXXXXX'
consumer_secret = 'XXXXXXXXXXXXXXXXXXXXXX'
access_token = 'XXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXX'
access_secret = 'XXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXX'

auth = tweepy.OAuthHandler(consumer_key, consumer_secret)
auth.set_access_token(access_token, access_secret)
api = tweepy.API(auth_handler = auth,
                parser = tweepy.parsers.JSONParser(),
                wait_on_rate_limit = True,
                wait_on_rate_limit_notify = True)
```

```python
#Write Json data of tweets to tweet_json.txt
import time
start_time = time.time()

with open('tweet_json.txt', 'w') as file:
    for tweet_id in df_twitter_arc['tweet_id']:
        try:
            tweet = api.get_status(tweet_id, tweet_mode='extended')
            file.write(json.dumps(tweet) + '\n') #writing one tweet per line
        except Exception as e:
            print("No tweet found for {} with error message {}".format(str(tweet_id), str(e)))
end_time = time.time()
print("Whole process finished in {} seconds".format(end_time - start_time))
```

```python
#Store data from JSON to dataframe
selected_attr = []
with open('tweet_json.txt', 'r') as json_file:
    for line in json_file:
        json_data = json.loads(line)
        selected_attr.append({'tweet_id': json_data['id'],
                              'favorites': json_data['favorite_count'],
                              'retweets': json_data['retweet_count'],
                              'timestamp': json_data['created_at'] })
additional_attrib = pd.DataFrame(selected_attr, columns=['tweet_id', 'favorites', 'retweets', 'timestamp'])
```
